```python
#Tickets Transactions sub task d
class Ticket:
    def __init__(self,c_id):
        self.cursor = mydb.cursor()
        self.c_id=c_id
        self.df=pd.read_sql_query("Select * from Conference",mydb)
    def isAvailable(self):
        print("----------------------------Number of Tickets Available----------------------------")
        available=df[df['Conference_ID']==self.c_id][['Conference_ID','Conference_Name','Tickets_Available']]
        print(available)
        if available.iloc[0,2] == 0:
            return False
        else:
            return True
    def ticketPurchasing(self):
        name,mail_id=[],[]
        ticket=int(input("Enter Number of Tickets needed (Maximum 2) : "))
        available=df[df['Conference_ID']==self.c_id][['Conference_ID','Conference_Name','Tickets_Available','Tickets_Purchased']]
        available_tick= available.iloc[0,2]
        purchase=available.iloc[0,3]
        if ticket <= 2:
            available_tick -= ticket
            purchase+=ticket
            for i in range(ticket):
                name.append(input("Enter name : "))
                mail_id.append(input("Enter Mail-ID : "))
            self.cursor.execute(" UPDATE Conference SET Tickets_Available = %s WHERE Conference_ID = %s" , (str(available_tick),str(self.c_id)))
            self.cursor.execute(" UPDATE Conference SET Tickets_Purchased = %s WHERE Conference_ID = %s" , (str(purchase),str(self.c_id)))
            mydb.commit()
            
            print("%d Ticket Received" % (ticket))
        else:
            print("You can't buy more than 2 Tickets")
```

```python
class Cancellation:
    def __init__(self,c_id):
        self.cursor=mydb.cursor()
        self.c_id=c_id
        self.df=pd.read_sql_query("Select * from Conference",mydb)
    def cancelTicket(self):
        available=self.df[df['Conference_ID']==self.c_id][['Conference_ID','Conference_Name','Conference_Date','Tickets_Available','Tickets_Purchased']]
        print(available)
        dat=available.iloc[0,2]
        available_tick= available.iloc[0,3]
        purchase=available.iloc[0,4]
        if purchase == 0:
            print("No, Tickets purchased so far")
            return
        if dat==str(date.today()):
            print("Cannot Cancel the Ticket")
            return
        canc=int(input("Enter the Number of Tickets to be Cancelled : "))
        available_tick += canc
        purchase-=canc
        self.cursor.execute(" UPDATE Conference SET Tickets_Available = %s WHERE Conference_ID = %s" , (str(available_tick),str(self.c_id)))
        self.cursor.execute(" UPDATE Conference SET Tickets_Purchased = %s WHERE Conference_ID = %s" , (str(purchase),str(self.c_id)))
        mydb.commit()
        print("%d Tickets Cancelled" % (canc))
```
